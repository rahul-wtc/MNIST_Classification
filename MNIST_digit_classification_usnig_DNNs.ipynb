```
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```
mnist =tf.keras.datasets.mnist
(x_train, y_train),(x_test, y_test) = mnist.load_data()
```

```
np.set_printoptions(threshold= np.inf, linewidth=280)
print(x_train[59999])
plt.imshow(x_train[59999])
```

```
#normalizing the trainig set from 0 to 1 for better calculation
x_train, y_train = x_train/255.0, y_train/255.0
```

```
#making model by creating layes
model = tf.keras.models.Sequential([
                                    tf.keras.layers.Flatten(input_shape = (28,28)),
                                    tf.keras.layers.Dense(256,activation = tf.nn.relu),
                                    tf.keras.layers.Dense(10, activation=tf.nn.softmax)
])
```

```
#Now set the optimizer and loss function for bachpropogation
model.compile(optimizer='adam', 
              loss ='SparseCategoricalCrossentropy',
              metrics = ['accuracy']
)
```

```
#training of model by fiting the given data
history = model.fit(x_train,y_train, epochs=10)
```

```
tf.keras.losses
```

```
plt.plot(history.history['loss'])
```

```
model.evaluate(x_test, y_test)
```
